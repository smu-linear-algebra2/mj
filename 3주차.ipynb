```python
import numpy as np
a=np.arange(12)
#ndarray.copy 메소드를 사용하면 새로운 numpy 객체의 새로운 메모리 공간에 배열 a의 데이터를 복사
b=a.copy()
#a와b는 같은 데이터를 갖고 있다.
print(a==b)
#하지만 a와 b는 서로 다른 numpy 배열 객체
print(a is b)
#b는 a의 데이터를 공유하지 않는다
print(b.base is a)
#따라서 데이터를 변경해도 서로 영향을 주지 않는다
a[0]=-1
b[11]=10000
print(a)
print(b)
```

```python
import numpy as np
import matplotlib.pyplot as plt
A=np.array([[7,2],[-7,5]])
b=np.array([-5,12])
x=np.linalg.solve(A,b)
#해벡터인 x를 그래프로 표현하기 위해 x1,x2로 표기한다
x1,y1=x
print(x1,y1)
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
#선형방정식을 직선으로 그린다
a1=A[:,0]
b1=A[:,1]
c1=-b
#c1x+c2y=c3
#c2y=c3-c1x
#y=(c3-c1x)/c2
for c1,c2,c3 in zip(A[:,0],A[:,1],b):
    x=np.linspace(-7,7,100)
    y=(c3-c1*x)/c2
    ax.plot(x,y,color='black')
#연립방정식의 해를 빨간 점으로 표시한다
ax.plot(x1,y1,'ro')
ax.axis([-7,7,-7,7])
ax.set_xticks(range(-7,7))
ax.set_yticks(range(-7,7))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspext('auto', adjustable='box')
ax.spines['left'].set_position('zerp')
ax.spines['bottom'].set_position('zerp')
ax.spines['left'].set_position('zerp')
ax.spines['left'].set_position('zerp')
```
